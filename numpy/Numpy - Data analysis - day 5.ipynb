```python
# 2 - Pyplot example
```

```python
a = np.arange(0, 6, 0.1)
```

```python
b1 = np.sin(a)
```

```python
b2 = np.cos(a)
```

```python
plt.plot(a, b1, label="sin")
```

```python
plt.plot(a, b2, linestyle="--", label="cos")
```

```python
plt.xlabel("a") # a axis name
```

```python
plt.ylabel("b") # b axis name
```

```python
plt.title("sin & cos")
```

```python
plt.legend()
```

```python
plt.show()
```
